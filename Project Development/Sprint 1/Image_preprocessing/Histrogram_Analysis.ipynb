```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
```

```python
image1 = imread('/content/23ea1618-d554-47fb-bc03-a1b978f14fbf___RS_HL 6008.JPG')
image2 = imread('/content/0a285c8b-1c31-48d4-89f2-af8b9edc36f6___RS_HL 5759.JPG')
```

```python
plt.figure()
plt.subplot(121), plt.imshow(image1)
plt.subplot(122), plt.imshow(image2)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.abs(image1[:, :, 0].astype(float) - image2[:, :, 0].astype(float)), cmap='gray')
plt.show()
```

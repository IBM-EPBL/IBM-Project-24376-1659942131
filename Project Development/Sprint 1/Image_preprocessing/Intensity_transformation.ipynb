# **Image Arithmetics**

```python
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np
```

```python
live = imread('/content/0b8dabb7-5f1b-4fdc-b3fa-30b289707b90___JR_FrgE.S 3047.JPG')
mask = imread('/content/0ebea6f4-08e4-4380-86f8-34d854697e32___JR_FrgE.S 2877.JPG')

plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(live, cmap='gray')
plt.subplot(122), plt.imshow(mask, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(live, cmap='gray')
plt.subplot(122), plt.imshow(live - 20, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(131), plt.imshow(mask - live, cmap='gray')
plt.subplot(132), plt.imshow(-(mask - live + 128), cmap='gray')
plt.subplot(133), plt.imshow(mask - live + 128, cmap='gray')
plt.show()
```

```python
shaded = imread('/content/0b8dabb7-5f1b-4fdc-b3fa-30b289707b90___JR_FrgE.S 3047.JPG')
shading = imread('/content/0ebea6f4-08e4-4380-86f8-34d854697e32___JR_FrgE.S 2877.JPG')

plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(shaded, cmap='gray')
plt.subplot(122), plt.imshow(shading, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(np.multiply(shaded, 1/shading), cmap='gray')
plt.subplot(122), plt.imshow(shaded, cmap='gray')
plt.show()
```

```python
xray = imread('/content/0b8dabb7-5f1b-4fdc-b3fa-30b289707b90___JR_FrgE.S 3047.JPG')
mask_xray = imread('/content/0a71ba91-8bde-4a91-84a9-bf950488f718___Rutg._Bact.S 2121.JPG')
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(xray, cmap='gray')
plt.subplot(122), plt.imshow(mask_xray, cmap='gray')
plt.show()
```

```python
plt.figure()
plt.imshow(np.multiply(xray, mask_xray/255), cmap='gray')
plt.show()
```
